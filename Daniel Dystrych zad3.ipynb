```python
import pandas as pd
import matplotlib.pyplot as plt
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
dane = pd.read_csv(url)
dane=dane.groupby(['Country/Region']).sum()
dane=dane.drop(['Lat','Long'],1)
```

```python
kraje=('Poland','Austria','Italy','Spain','US','Australia')
colors=('blue','red','orange','green','black','violet')
x=pd.to_datetime(dane.loc['Poland'].index)
plt.rcParams["figure.figsize"] = (12,6) 
for i in range(6):
    plt.plot(pd.to_datetime(dane.loc['Poland'].index),dane.loc[kraje[i]].values,color=colors[i])
    plt.grid()
    plt.title('Koronawirus in '+kraje[i])
    plt.show()

for i in range(6):
    plt.plot(pd.to_datetime(dane.loc['Poland'].index),dane.loc[kraje[i]].values,color=colors[i])
    plt.grid()
plt.legend(kraje)
plt.grid()
plt.yscale('log')
plt.title('Zachorowania na wiecie [skala log]')
plt.show()
```

Patrzc na powysze wykresy mona zauway, e w przypadku niektrych krajw nastpuje przegicie krzywej (Australia, Austria, Wochy i Hiszpania). Mona wic wyciagn wnioski e nastpi ju u nich szczyt przyrostu zachorowa. 
Mona zauway te z powyszych wykresw, e epidemia ma na pocztku prawie zawsze wzrost wykadniczy, jednake niektre kraje takie jak Polska i Stany bardzo szybko przechodz w wzorst liniowy. Moe to by spowodowane za pewne szybkimi dziaaniami rzdu i ograniczeniami nakadanymi na spoeczestwo ('Polska') lub du powierzchni kraju i liczb ludnoci ktre nie pozwalaj na bardzo szybki przyrost liczby zachorowa.
