```python
import pandas as pd
import matplotlib.pyplot as plt
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
dane = pd.read_csv(url)
dane=dane.groupby(['Country/Region']).sum()
dane=dane.drop(['Lat','Long'],1)
```

```python
kraje=('Poland','Austria','Italy','Spain','US','Australia')
colors=('blue','red','orange','green','black','violet')
x=pd.to_datetime(dane.loc['Poland'].index)
plt.rcParams["figure.figsize"] = (12,6) 
for i in range(6):
    plt.plot(pd.to_datetime(dane.loc['Poland'].index),dane.loc[kraje[i]].values,color=colors[i])
    plt.grid()
    plt.title('Koronawirus in '+kraje[i])
    plt.show()

for i in range(6):
    plt.plot(pd.to_datetime(dane.loc['Poland'].index),dane.loc[kraje[i]].values,color=colors[i])
    plt.grid()
plt.legend(kraje)
plt.grid()
plt.yscale('log')
plt.title('Zachorowania na świecie [skala log]')
plt.show()
```

Patrząc na powyższe wykresy można zauważyć, że w przypadku niektórych krajów następuje przegięcie krzywej (Australia, Austria, Włochy i Hiszpania). Można więc wyciagnąć wnioski że nastąpił już u nich szczyt przyrostu zachorowań. 
Można zauważyć też z powyższych wykresów, że epidemia ma na początku prawie zawsze wzrost wykładniczy, jednakże niektóre kraje takie jak Polska i Stany bardzo szybko przechodzą w wzorst liniowy. Może to być spowodowane za pewne szybkimi działaniami rządu i ograniczeniami nakładanymi na społeczeństwo ('Polska') lub dużą powierzchnią kraju i liczbą ludności które nie pozwalają na bardzo szybki przyrost liczby zachorowań.
